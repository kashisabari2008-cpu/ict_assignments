QUESTION 1

```python
books = {}
issued = []

while True:
    print("\n1.Add 2.Issue 3.Return 4.Search 5.Exit")
    ch = input("Choice: ")

    if ch == "1":
        b = input("Book ID: ")
        books[b] = [input("Title: "), int(input("Copies: "))]
        print("Added")

    elif ch == "2":
        s = input("Student ID: ")
        b = input("Book ID: ")
        if b in books and books[b][1] > 0:
            issued.append((s, b))
            books[b][1] -= 1
            print("Issued")
        else:
            print("Not available")

    elif ch == "3":
        s = input("Student ID: ")
        b = input("Book ID: ")
        if (s, b) in issued:
            issued.remove((s, b))
            books[b][1] += 1
            print("Returned")
        else:
            print("Record not found")

    elif ch == "4":
        k = input("Search: ").lower()
        for i in books:`
            if k in books[i][0].lower():
                print(i, books[i])

    elif ch == "5":
        break

    else:
        print("Invalid")
```

QUESTION 2

```python
s = {}

while True:
    print("\n1.Add 2.Marks 3.Report 4.Exit")
    c = input("Choice: ")

    if c == "1":
        r = input("Roll ni: ")
        s[r] = [input("Name: "), [0,0,0]]

    elif c == "2":
        r = input("Roll no: ")
        s[r][1] = [int(input("M: ")), int(input("S: ")), int(input("E: "))]

    elif c == "3":
        for r in s:
            a = sum(s[r][1]) / 3
            print(r, s[r][0], "PASS" if a >= 40 else "FAIL", a)

    elif c == "4":
        break
```
